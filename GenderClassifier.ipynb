# Importing Data and conversion to arrays

```python
from os import listdir
from os.path import isfile, join
from sklearn.metrics import accuracy_score
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import ndimage
from time import time


# Data , add your own path , male folder includes male images , female folder for female images (scraped from google using chrome extension)
maleFolder = 'testData/male'
femaleFolder = 'testData/female'

malePhotos = [join(maleFolder, f) for f in listdir(maleFolder) if isfile(join(maleFolder, f))]
femalePhotos = [join(femaleFolder, f) for f in listdir(femaleFolder) if isfile(join(femaleFolder, f))]
malePhotos.remove('testData/male/.DS_Store')
femalePhotos.remove('testData/female/.DS_Store')

print("Male photos:- ",len(malePhotos) ,"\nFemale Photos:- " ,len(femalePhotos))
# COrrectly Printing

maleFrame = [[x for x in malePhotos] , [0 for i in range(0,len(malePhotos))]]
femaleFrame = [[x for x in femalePhotos] , [1 for i in range(0,len(femalePhotos))]]

frame = {
    'photo' : malePhotos + femalePhotos ,
    'gender' : [0 for i in range(0,len(malePhotos))] + [1 for i in range(0,len(femalePhotos))]
}

data = pd.DataFrame.from_dict(frame)


#print(data.head())

# Made the dataframe COrrectly

imgarr = np.ndarray(shape = (len(data['photo']),160,160),dtype=np.float32)
from scipy.misc import imresize
x = 0
print("Converting Photos to array:-")
for i in data['photo']:
    
    img = (ndimage.imread(i,flatten = True).astype(float) - (255.0/2))/255.0
    #print((img.shape))
    img = imresize(img,[160,160])
    #print(img.shape)
    if img.shape != (160,160): 
        x+=1
        continue
    imgarr[x,:,:] = img
    #print(imgarr[x].shape)
    x += 1
# Converted the photos into nd array
#imgarr = imgarr[:200,:,:]
imgarr = imgarr.reshape(imgarr.shape[0],160*160)

features = imgarr
labels = data['gender']
print('Success')
```

## Cross Validation (Done on different algos and Gradient Boosting selected for training and testing)

## Final Training and Testing

```python
from sklearn.ensemble import GradientBoostingClassifier
clf = GradientBoostingClassifier()

t0 = time()
clf.fit(features,labels)
t2 = time()
print("Training Time:- " , t2-t0)
```

## Saving the model

```python

import pickle
clf = pickle.load(open('train_fest.pkl','rb'))
def gender(i):
    i = imresize(i,[160,160])
    i=i.reshape(160*160)
    return (clf.predict(i))
```

```python
import cv2
import numpy as np
font =  cv2.FONT_HERSHEY_SIMPLEX
faceDetect = cv2.CascadeClassifier('haarcascade_frontalface_alt2.xml')
cam = cv2.VideoCapture(0)
#cam = cv2.VideoCapture('../faces.mp4')
w = cam.get(3)
w = int(w)
h = cam.get(4)
h = int(h)

fourcc = cv2.VideoWriter_fourcc(*'mp4v')
#out = cv2.VideoWriter('../output.m4v',fourcc, 5.0, (w,h))
male = 0
female = 0
while(True):
    ret,img = cam.read()
    
    gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
    faces = faceDetect.detectMultiScale(gray,1.15,5)
    for(x,y,w,h) in faces:
        gen = gender(gray[y:y+h+40 , x:x+w+40])
        #gen = 0
        if gen == -1: continue
        if gen==0: 
            gen = "Male"
            male+=1
        else: 
            gen = "Female"
            female+=1
        cv2.putText(img,gen,(x,y-10), font, 2,(0,0,255),2,cv2.LINE_AA)
        cv2.rectangle(img,(x,y),(x+w,y+h),(0,255,0),2)
    cv2.imshow('Face',img)
    #if ret : out.write(img)
    if cv2.waitKey(1) & 0xFF == ord('q'): 
        break
    
cam.release()
#out.release()
cv2.destroyAllWindows()
```
